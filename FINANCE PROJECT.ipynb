```python
KOTAKMAHINDRA = data.DataReader('KOTAKBANK.NS','yahoo',start='2011-01-01',end='2019-01-01')
```

```python
tickers = ['AXIS','HDFCBANK','ICICI','KOTAKMAHINDRA','SBI']
```

```python
pd.concat([AXIS,HDFCBANK,ICICI,KOTAKMAHINDRA,SBI],keys=tickers,axis=1)
```

```python
bank_stocks = pd.concat([AXIS,HDFCBANK,ICICI,KOTAKMAHINDRA,SBI],keys=tickers,axis=1)
```

```python
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max() ##very important
```
